## 1.2 URL 생성 및 조회

### < URL 구조>

url = 기본주소?항목A=값1&항목B=값2&항목3=값3...... 

ex) http://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001&date=20171026&page=3
   - 기본주소 : http://news.naver.com/main/list.nhn   : 네이버뉴스속보 메인
   - mode=LSD :  ?? 없어도 조회가능함?
   - mid=sec :  sec 인 경우 속보. 카테고리가 정해진 경우는 shm
   - sid1 = 001 : 속보는 001, 세계는 104, 경제는 101...
   - date = 20171026 : 조회일자
   - page = 3 :  3번째 페이지를 검색
   

### < URL 주소 만드는 법>

1. 텍스트 작업
2. urllib.parse 활용
3. requests params 속성 활용

```python
t1='가가가가가가'
t2='sksksksksks'
print(t1+t2)
name='장남수'
names=['홍길동', '장남수','천하장사']

for name in names:
    t3="내이름은 {} 입니다".format(name)
    print(t3)
```

```python
# 1. 텍스트로 만들기

base_url = 'http://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001&date={}&page={}'
date=20171026
page=5

dates=[1,2,3,4,5]

for date in dates:
    url = base_url.format(date,page)
    print(url)


```

```python
base_url = 'http://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001&date='
date = '20171102'
page='123'

url = base_url+date+"&page="+page

print(url)
```

```python
# 2. urllib.parse 활용

import urllib.parse

base_url = 'http://news.naver.com/main/list.nhn'

#속성 값을 dict 형태로 저장
values = {
    'mode':'LSD',
    'mid':'sec',
    'sid1':'001',
    'date':'date',
    'page':'page'
}

#urlencode 통해 속성정보 합치기
params = urllib.parse.urlencode(values)
print(params)

url = base_url+'?'+params
print(url)
```

```python
# 3. requests  params 속성 활용하기

import requests

base_url = 'http://news.naver.com/main/list.nhn'
values = {
    'mode':'LSD',
    'mid':'sec',
    'sid1':'001',
    'date':'date',
    'page':'page'
}

#url 기본주소와 속성 정보(dict형태)로 원하는 주소에 바로 접근
data =  requests.get(base_url, params = values)

print(data.url)

print(data.status_code)   # 200은 url 주소 접근완료. 
print(data)
#data.text  or data.content 로 페이지내용 확인 가능함
```

## 1.3 내용 가져오기

```python
import requests
from bs4 import BeautifulSoup

url = 'https://github.com/namsoojang/Study_Play_with_data'

#url 페이지 내용 가져오기
# requests.get(url).text: 텍스트   requests.get(url).content:이미지(바이너리 자료형)
html = requests.get(url).text              
# print(html)


# html으로 작업하기 편하도록 BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
# print(soup)

print(type(soup))
```
